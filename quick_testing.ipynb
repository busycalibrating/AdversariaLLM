```python
import torch
import torch.optim as optim

# Gaussian function definition
def gaussian_2d(xy, mean, cov):
    """
    xy: torch.Tensor of shape (n_points, 2) - the points in 2D space.
    mean: torch.Tensor of shape (2,) - mean of the Gaussian.
    cov: torch.Tensor of shape (2, 2) - covariance matrix of the Gaussian.
    """
    diff = xy - mean
    inv_cov = torch.inverse(cov + cov.T)
    exponent = -0.5 * torch.sum(diff @ inv_cov * diff, dim=1)
    return torch.exp(exponent) / (2 * torch.pi * torch.sqrt(torch.det(cov)))

# Data: Randomly generated points in 2D space (you can replace this with your actual data)
torch.manual_seed(0)
data = torch.tensor([[0.5, 1.2], [1.3, 1.1], [0.9, 1.5], [1.2, 0.9], [1.0, 1.3]], dtype=torch.float32)

# Initial guesses for parameters: mean and covariance matrix
mean = torch.tensor([1.0, 1.0], requires_grad=True)  # Initial guess for mean
cov = torch.eye(2, requires_grad=True) # Initial guess for covariance matrix (identity matrix)

# Optimizer
optimizer = optim.Adam([mean, cov], lr=0.01)

# Training loop
for step in range(1000):  # Number of optimization steps
    optimizer.zero_grad()

    # Compute Gaussian values for the current mean and covariance
    gaussian_values = gaussian_2d(data, mean, cov)

    # Loss: We want the Gaussian to fit the data points, so we minimize the negative log-likelihood
    loss = -torch.sum(torch.log(gaussian_values))

    # Backpropagation
    loss.backward()

    # Optimization step
    optimizer.step()

    if step % 100 == 0:
        print(f'Step {step}, Loss: {loss.item()}, Mean: {mean.data}, Covariance: {cov.data}')

# Final fitted parameters
print(f'Fitted Mean: {mean.data}')
print(f'Fitted Covariance: {cov.data}')
```
